```python
import pandas as pd
```

```python
# load Data
df = pd.read_csv(r'C:\Users\Batna\Documents\InternShip\Data\sales_data_sample.csv', encoding='latin1')

df.head()
print("Shape:", df.shape)
print("Rows:", len(df))
print("Columns:", df.shape[1])
print("Column names:", df.columns.tolist())
df.info()
```

```python
# check for missing values
df.isnull().sum()
```

```python
df['STATE'] = df.groupby('CITY')['STATE'].transform(lambda x: x.fillna(x.mode()[0]) if not x.mode().empty else 'N/A')
df['TERRITORY'] =df.groupby('COUNTRY')['TERRITORY'].transform(lambda x: x.fillna(x.mode() [0]) if not x.mode().empty else 'N/A')
df['POSTALCODE'] =df.groupby('CITY') ['POSTALCODE'].transform(lambda x: x.fillna(x.mode() [0] )if not x.mode().empty else 'N/A')
df['TERRITORY'] = df.groupby('STATE')['TERRITORY'].transform(lambda x: x.fillna(x.mode()[0]) if not x.mode().empty else x)
df['TERRITORY'] = df.groupby('POSTALCODE')['TERRITORY'].transform(lambda x: x.fillna(x.mode()[0]) if not x.mode().empty else 'N/A')
#for group_col in ['COUNTRY','STATE','POSTALCODE']: df['TERRITORY']=df.groupby('group_col') ['TERRITORY'].transform(lambda x: x.fillna(x.mode()[0]) if not x.mode().empty else 'N/A')
```

```python
# check for missing values
df.isnull().sum()
```

```python
df = df.drop('ADDRESSLINE2', axis=1)
print("Shape:", df.shape)
```

```python
print(df.columns.tolist())
```

```python
df.head(50)
```

```python
#Remove Dupicates
df.drop_duplicates(inplace=True)
```

```python
#Standardize Text and Clean Columns
# Standardize column names
df.columns = [col.lower().replace(' ', '_') for col in df.columns]

# Example: Strip and lowercase all string columns
for col in df.select_dtypes(include='object').columns:
    df[col] = df[col].astype(str).str.strip().str.lower()
```

```python
print(df.columns.tolist())
```

```python
#Convert Date Columns
#df['orderdate'] = pd.to_datetime(df['orderdate'], errors='coerce', dayfirst=True)
df['orderdate'] = pd.to_datetime(df['orderdate'], errors='coerce', dayfirst=False)
```
